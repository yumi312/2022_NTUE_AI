```python
#WHO的數據來由IBM或微軟的專業團隊管理，通常非常乾淨，沒有太多資料純化的工作要做。
#1. "dateRep"是日期欄位，但未被視系統為datetime，並且按從最新到最早的順序排序。如果不進行合理化，會在以後進行運算時發生問題。
#日期格式問題：將“ dateRep”列轉換為日期時間,並對每個國家的數據進行重新排序按時間順序，重新索引數據以防止在創建聚合(aggregates)時出錯。
#2. 排序用法：Data.sort.value
# axis 参數的初始值為0，是和index撘配使用，axiz=0跨行排序，axis=1，columns，跨列排序
# by参數要求传傳入一個字串或者是一個字串列表，用來指定按照axis的中的哪個元素来排序
# scending参數的初始值是True，按照升序排序，設為False，按照降序排列

#kind参數：排序算法，初始值是quicksort（快速排序），mergesort（歸並排序）或者是heapsort（堆排序）
#inplace参數：In-place algorithm（原地算法），初始值為0,運算後不存入原稿，設為1,存入原稿。
#na_position:定義DataFrame中的空缺值的順序，初始值last表示將空缺值放在排序的最後一個，first放在最前。

#pandas.Series也有sort_values方法，但是和Dataframe.sort_values的用法很接近
```

```python
#6-2: 資料純化及排序：為了該使用者有節奏性的互動感覺，故先使用年月日順序進行排序。
```

```python
display(data.head())
#                                                 DD/MM/YYYY 
#                 轉換日期時間格式                 14/12/2020    ->   2020-12-14
data['dateRep'] = pd.to_datetime(data['dateRep'], dayfirst=True) #將欄位「DateRep」格式轉換datatime格式
data.sort_values(by=['dateRep','countriesAndTerritories'],       #按日期及國家碼進行排序
                 ascending=True, inplace=True)
display(data.head())
data = data.reindex() #排序結果放回Data中
df1=data
display(df1.head())
```

```python
#6-3:進行專業視覺的資料顯示：
#(1) 特定資料著色進行視覺強化：ighlight_null('') 
#(2) 資料旁加條狀圖：bar('deaths', vmin=0,color='gray') 
#(3) 篩選資料著色進行數字整理：format('{:.0f}', subset='cases') 
#(4) 特定欄位排序後用漸層色強化：
#(5) background_gradient('Reds', subset='popData2019') 
```

```python
# 資料查詢 Dataframe Query
#               字串資料裡面包含單引號，通常會需要Escape character \'才能正確顯示單引號，但雙引號不用
df1 = data.query("dateRep == '05/10/2020'")  # == df1 = data[data['dataRep'] == '05/10/2020']
df2 = df1.query("cases > 100")
df3 = df2.sort_values(by=['cases','deaths'],ascending=False)
#依二個欄位為指標進行排序並降冪(大到小)排序
df3.head()
```

```python
df3.info()
df3.describe()
# len(df3)
```

```python
import datetime
#                    小數後顯示到第幾位，.0f沒有小數位數，四捨五入
print(f'共 {len(df3):.0f} 個國家在指定日期，單日確診數超過100人') #數字顯示設定

title1='★☆★☆== 今日('+str(df3.iloc[1].dateRep)+')全球Covid－19確診人數超過100人之國家，共有'+str(len(df3))+'個 ==☆★☆★'

# (pandas)類別，html
(df3.style
     .format('{:.0f}', subset='cases')                          #將欄位的數值顯示限制到小數後第幾位
     .format('{:.0f}', subset='deaths')                         #將欄位的數值顯示限制到小數後第幾位
     .format('{:.0f}', subset='popData2019')                    #將欄位的數值顯示限制到小數後第幾位

     .highlight_max(subset=['cases'],color='pink')              #標示欄位最大值並用底色強化
     .set_caption(title1) #添加標題
     .hide_index()                                              #隱藏索引欄（最左邊）
 
     .bar('deaths', vmin=0,color='gray')                        #將欄位依數值大小畫條狀圖(黑色)
     .highlight_max('cases',color='lightblue')                  #標示出欄位最大值並底色強化
     .highlight_min('deaths',color='red')                       #標示出特定欄位最大值並底色強化
     .background_gradient('Reds', subset='popData2019')         #將欄位依數值大小在背景著色  漸層
     .highlight_null('')                                        #將空值顯示底色為紅色
)
```

```python
#6-5. 較複雜的篩選和排序：
#這𥚃有用一些新的做法，將列出的數據特別設定一位小數及二位小數。這種用法很好用，直覺式的設計，是2020年以後資料工程的必備能力。
#(1) 台灣Covid－19死亡紀錄：篩選出台灣死亡病例的發生時間及人數。
```

```python
#台灣Covid－19死亡紀錄：篩選出台灣死亡病例的發生時間及人數
data_filter = (data.countriesAndTerritories == 'Taiwan') & (data.deaths > 0)
data_filter
```

```python
(data[data_filter]
    .style
    .format('{:.1f}', subset='deaths')                 #將欄位的數值顯示限制到小數後第一位
    .format('{:.2f}', subset='popData2019')            #將欄位的數值顯示限制到小數後第二位
    .applymap(lambda x: 'background-color: rgb(233, 160, 215)', #強化視覺在二個特定欄位加底色
              subset=pd.IndexSlice[:, ['deaths','popData2019']]))
```

```python
#美國Covid－19死亡紀錄：篩選「死亡病例大於3000人」的發生時間及人數
data_filter = (data.countriesAndTerritories == 'United_States_of_America') & (data.deaths > 3000)
data_filter
(data[data_filter]
    .style
    .format('{:.0f}', subset='deaths')                            #將欄位的數值顯示限制到小數後第一位
    .format('{:.0f}', subset='popData2019')                       #將欄位的數值顯示限制到小數後第二位
    .applymap(lambda x: 'background-color: rgb(233, 190, 215)', 
              subset=pd.IndexSlice[:,'cases':'deaths']))
```

```python
#中國Covid－19 確診及死亡紀錄：篩選「死亡病例大於100人」的發生時間及人數
#從資料上看中國從2020/4/17 之後就蓋牌了。可從 2020/4/10-2020/4/30 得到佐證

data_filter = (data.countriesAndTerritories == 'China') & (data.deaths > 120)
data_filter
(data[data_filter]
    .style
    .format('{:.1f}', subset='deaths')                            #將欄位的數值顯示限制到小數後第一位
    .format('{:.0f}', subset='popData2019')                       #將欄位的數值顯示限制到小數後第二位
    .applymap(lambda x: 'background-color: rgb(133, 160, 215)', 
              subset=pd.IndexSlice[:, ['deaths','popData2019']]))
```

```python
#中國從2020/4/17 之後就大量減少。從 2020/4/10-2020/4/30得到佐證。

data_filter = (data.countriesAndTerritories == 'China') & (data.month ==4) & (data.day >16)
data_filter
(data[data_filter]
    .style
    .format('{:.1f}', subset='deaths')                        #將欄位的數值顯示限制到小數後第一位
    .format('{:.1f}', subset='popData2019')                   #將欄位的數值顯示限制到小數後第一位
    .applymap(lambda x: 'background-color: rgb(233, 160, 215)', 
              subset=pd.IndexSlice[:, 'deaths']))
```

```python
#6-6: 累計各國Covid-19到今天為止的確診人數，及確診天數
display(data.groupby(['countriesAndTerritories']).agg({'cases': ['sum', 'count']}))
```

```python
# 6-7. 群組資料(groupby):對五大洲各國家分別進行Covid-19計算，同時做樞紐分析(矩陣式)：每日平均值 和 累計總數 和發生天數(有確診例發生的天數)
# ALL：中間是平均每個國家每天確診數，全球總數，及總發生天數（加總每個國家有確診的總天數，這個數字沒有統計上的意義）
df_cases = data.pivot_table(index=['countriesAndTerritories','continentExp'], values='cases', margins=True, aggfunc=['mean', 'sum','count'])
df_cases
```
