```python
import plotly.express as px
import pandas as pd
import datetime

print("資料擷取，AI運算中.....")
import time
tStart = time.time()#計時開始

#data = pd.read_csv('https://opendata.ecdc.europa.eu/covid19/casedistribution/csv',usecols=list(range(0,11)))
#data.to_csv('Covid19/WHO.csv') 

data = pd.read_csv('Covid19/WHO.csv') 

Saved_Name='Covid19/WHO.csv'

print("完成了",end='')
tEnd = time.time()#計時結束
print(f'{len(data):.0f} 筆資料,{(tEnd - tStart):.2f}秒') #數字顯示規格設定

import os
print("本圖存檔位置:"+str(os.getcwd())+str("/")+str(Saved_Name)) #取得當前路徑                                                
print("共有多少國家納入WHO統計："+str(data['countriesAndTerritories'].nunique()))  #找出相異資料
print("共統計多少天："+str(data['dateRep'].nunique()))                     #找出相異資料

                                                                        
#data.year = df.year.fillna((data.year.shift() + data.year.shift(-1))/2) #資料純化：把NaN資料替換成0
data.head()
```

```python
# 對五大洲各國家分別進行Covid-19計算，同時做樞紐分析(矩陣式)：每日平均值 和 累計總數 和發生天數(有確診例發生的天數)
# ALL：中間是平均每個國家每天確診數，全球總數，及總發生天數（加總每個國家有確診的總天數，這個數字沒有統計上的意義）
df_cases = data.pivot_table(index=['countriesAndTerritories','continentExp'], values='cases', margins=True, aggfunc=['mean', 'sum','count'])
df_cases
```

```python
#Python地圖可視化庫有pyecharts、plotly、folium，，還有較不流行bokeh、basemap、geopandas，也是地圖可視化的利器。
#Plotly Express 是高級 Python 可視化庫：是 Plotly.py高級封裝，為複雜的圖表提供了一個簡單的語法。 
#受 Seaborn 和 ggplot2 的啟發，它專門設計為具有簡潔且易於學習的 API ：只要一次導入，可以在一個函數調用中創建豐富的交互式繪圖，
#包括分面繪圖（faceting）、地圖、動畫和趨勢線。 用 pip install plotly_express 命令可以安裝 Plotly Express。
```

```python
#df4 = data.query("continentExp == 'Oceania'")
df5 = data.sort_values(by=['year','month','day'],ascending=True)#依欄位為指標進行排序，並升冪(大到小)排序
#display(df5)
fig = px.choropleth(df5,
                    locations ='countryterritoryCode',
                    color='cases',
                    animation_frame="dateRep",
                    color_continuous_scale=px.colors.sequential.solar_r,#colors.diverging.RdBu,
                    projection = 'natural earth')
title2='2019/12/31 ~ '+str(datetime.date.today())+' (WHO日內瓦時間) 全球各國每日確診數'
fig.update_layout(title=title2,
                    font=dict(family="Arial, monospace",size=14,color="#2f2faf")
                 )

fig.show()
```
