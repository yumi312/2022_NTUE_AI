# 練習

- 觀察 http://www.boxofficemojo.com/yearly/ 並撰寫爬蟲程式
- 抓取每年度冠軍排行榜
- 使用 requests + BeautifulSoup 實作
- 透過 pandas 輸出成 csv

```python
# requests - 模擬送出需求給Server
# BeautifulSoup - 用來解析HTML語法的套件
```

```python
import os
import requests
import pandas as pd

from bs4 import BeautifulSoup

url = 'http://www.boxofficemojo.com/yearly/'
```

```python
resp = requests.get(url)
resp.encoding = 'utf-8'
resp.text
```

```python
#BeautifulSoup(HTML程式碼, 'html.parser'/'lxml(處理html和xml)')
soup = BeautifulSoup(resp.text, 'lxml')
```

```python
# table = soup.find('table', attrs={'cellspacing': '1'})
# print(table)
```

```python
table = soup.find('div', {'id': 'table'})
print(table)
```

```python
rows = table.find_all('tr')
rows
```

```python
from pprint import pprint

# rows 為List 可用pop取出資料
colname = rows.pop(0)
colname
```

```python
colname = [i.text for i in colname]
colname
```

```python
#stripped_strings:自動去除空白字串
rows = [list(row.stripped_strings) for row in rows]
rows
```

```python
df = pd.DataFrame(rows, columns=colname)
df
```

```python
results = os.path.abspath('/results')
if not os.path.exists(results):
    os.makedirs(results)

filename = os.path.join(results, 'boxofficemojo.csv')
df.to_csv(filename, index=False)
print('Save csv to {}'.format(filename))
```
