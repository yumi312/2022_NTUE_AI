- 透過 pandas 輸出成 csv

```python
import os
import requests
import pandas as pd

from bs4 import BeautifulSoup

url = 'https://rate.bot.com.tw/xrt?Lang=zh-TW'
```

```python
resp = requests.get(url)
resp.encoding = 'utf-8'
resp.text
```

```python
#BeautifulSoup(HTML程式碼, 'html.parser'/'lxml(處理html和xml)')
soup = BeautifulSoup(resp.text, 'lxml')
```

```python
table = soup.find('table', {'title': '牌告匯率'})
print(table)
```

```python
rows = table.find_all('th',{"class":"display_none_print_show print_width"} )
rows
```

```python
from pprint import pprint

colname = rows
# print(type(colname))
colname
```

```python
colname = [i.string for i in colname]
colname
```

```python
colname.insert(0, soup.find("th",  {'class':"print_width set-title-L-min-width-class noscript"}).text)
colname
```

```python
cols = []
cols.append(colname[0])
for col in colname[1:3]:
    cols.append(col+colname[3]) 
    cols.append(col+colname[4]) 
print(cols)
```

```python
rows = table.find_all('tr')
rows
```

```python
rows = [list(row.stripped_strings) for row in rows]
rows
```

```python
rows.pop(0)
rows.pop(0)
```

```python
data = []
for row in rows:
    data.append(row[1:6])
data
```

```python
cols
```

```python
df = pd.DataFrame(data, columns=cols)
df
```
